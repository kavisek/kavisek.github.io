# Every Matplotlib Plot Linestyle

<span>Matplotlib has many lifestyles for its line plots. Sometimes is hard to visualize each lifestyle from just its line marker. Therefore I created a quick notebook to plot each line style out. I often come back to these plot to refresh my memory.</span>


linestyle documentation:  https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html 

### Import Preliminaries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

linestyles = ['-','--','-.',':']
```

### Visualize Matplotlib Markers

```python
for line in linestyles:
  data = pd.DataFrame(np.random.choice([1,2,3,4], replace=True, size=10), 
                      columns=['Col A'])
  plt.plot(data.index,data['Col A'], linestyle=line, 
           color='purple', linewidth=3)
  plt.title('"'+str(line)+'" '+'Line Style')
  plt.ylabel('Y Label')
  plt.xlabel('X Label')
  plt.show()
```
