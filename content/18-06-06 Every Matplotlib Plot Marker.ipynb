# Every Matplotlib Plot Marker

<span>Matplotlib has many markers for its line plots. Sometimes is hard to visualize each marker from just its line marker code. Therefore I created a quick notebook to plot each marker out. I often come back to these plot to refresh my memory on which marker is which.</span>

Marker documentation: https://matplotlib.org/api/markers_api.html 

### Implement Preliminaries

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'

# Import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# List of matplot lib markers
markers = ['.', ',', 'o', 'v', '^', '<', '>', '1',
           '2', '3', '4', '8', 's', 'p', 'P', '*', 'h',
           'H', '+', 'x', 'X', 'D', 'd', '|', '_', '$yolo $',
           'None']
```

### Visualize Matplotlib Markers

```python
# Plot a line for each marker in our list
for mark in markers:
    data = pd.DataFrame(np.random.choice([1,2,3,4], replace=True, size=10), 
                      columns=['Col A'])
    plt.plot(data.index,data['Col A'], marker=mark, 
           color='purple', markersize=12)
    plt.title('"'+str(mark)+'"'+" "+'Marker Plot')
    plt.ylabel('Y Label')
    plt.xlabel('X Label')
    plt.show()
```
