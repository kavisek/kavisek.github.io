# Hierarchical Clustering

<span>This example of hierarchical clustering is more for my amusement. I want to explore some hierarchical clustering coder with some data about pokemon from Kaggle. Let's see if we can cluster pokemon together using the Scipy and Matplotlib packages.</span>

### Import Prelimanries

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'

# Import modules
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib as mpl
import numpy as np
import pandas as pd 
import sklearn

from scipy.spatial.distance import pdist
from scipy.cluster import hierarchy

# Set pandas options
pd.set_option('max_columns',1000)
pd.set_option('max_rows',30)
pd.set_option('display.float_format', lambda x: '%.3f' % x)

# Set plotting options
mpl.rcParams['figure.figsize'] = (8.0, 7.0)
```

### Import Data

```python
# Import pokemon dataframe
df = pd.read_csv('Data/Pokemon/pokemon.csv', index_col='#')

# Filter dataframe for only the first Generatino
df = df[df.Generation == 1]

# Store the names of the pokemon in a seperate object
names = df.Name

# View the head of the datafram
df.head()
```

### Dendogram

```python
# Perform hierarchical/agglomerative clustering on the pokemon data using only
#  stat features of each pokemon using "Ward Variance Minimization Algorithm"
linkagemat = hierarchy.linkage(df.loc[:,'HP':'Speed'], method='ward')

# Plot the dendograms 
plt.figure(figsize=(10, 25))
hierarchy.set_link_color_palette(['#1B43AA','#02D7F5', '#98E2BC','#019CA4'])
hierarchy.dendrogram(
    linkagemat,
    leaf_rotation=0,
    leaf_font_size=8., 
    labels = names.values,
    orientation ='right',
    above_threshold_color='#81B3DB')                                          
plt.ylabel('Pokemon')
plt.xlabel('Distance')
plt.title('Dendogram of Pokemon by Stats Values');
```

```python
# Perform hierarchical/agglomerative clustering on the pokemon data using only
#  stat features of each pokemon using "Farthest Point Algorithm"
linkagemat = hierarchy.linkage(df.loc[:,'HP':'Speed'],method='average')

plt.figure(figsize=(10, 25))
hierarchy.set_link_color_palette(['#1B43AA','#02D7F5', '#98E2BC','#019CA4'])
hierarchy.dendrogram(
    linkagemat,
    leaf_rotation=0,
    leaf_font_size=8., 
    labels = names.values,
    orientation ='right',
    color_threshold=125,
    above_threshold_color='#81B3DB')                             
                                  
plt.ylabel('Pokemon')
plt.xlabel('Distance')
plt.title('Dendogram of Pokemon by Stats Values');
```

There are 5 other algorithims that you can use in your hierarchy clustering. The approaches can be found on Sklearn's website at the following link: https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.linkage.html
